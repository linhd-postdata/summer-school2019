### Variables

Las variables son contenedores de información donde podemos almacenar aquellos datos con los que trabajaremos posteriormente.

````
x = 3
````

Una vez almacenados podemos operar con ellos y ver el resultado con la función **`print()`** de Python:

````
print(x - 1)
print(x / 2)
print(x + 1)
print(x * x)
````

Podemos cambiar el valor de estas variables por otro cualquiera, e incluso actualizar dicho valor. Pongamos por ejemplo que tenemos una variable que representa el número de libros que tenemos en una biblioteca:

````
num_libros = 100
````

Si queremos incrementar el valor de esta variable podemos asignarle un nuevo valor que sea el que tiene actualmente, más 1:

````
num_libros = num_libros + 1
print(num_libros)
````

O también podemos almacenar el valor de esa variable en una nueva, para no modificar la original:

````
nuevos_libros = num_libros
nuevos_libros = nuevos_libros + 1
print(nuevos_libros)
# Comprobamos que el valor de la variable original no ha cambiado
print(num_libros)
````

Las variables también pueden almacenar otros tipos de datos como texto. A esto se le llama "cadena de texto" o "cadena" para abreviar. Cualquier valor que pongamos entre comillas simples o dobles lo estamos definiendo como una cadena. Si no ponemos comillas python pensará que estamos haciendo referencia a una variable definida anteriormente y probablemente falle al no encontrar ninguna con ese nombre:

````
author = Fray Luis de León
````

¡Arreglémoslo con la sintaxis correcta!

````
author = "Fray Luis de León"
````

Con este ejemplo hemos almacenado una cadena de texto (el nombre de un autor) en la variable **author**. Al igual que en el ejemplo anterior, podemos ver el valor de esta variable con la función **print()**:

````
print(author)
````

El valor de una variable se puede cambiar, lo que sobreescribirá el valor almacenado anteriormente.

````
author="Gloria Fuertes"
print(author)
````

Las cadenas son una sucesiñon de caracteres (letras, númeos, espacios, signos ortográficos, etc). Podemos acceder a cada uno de estos caracteres individuales con la ayuda de los `índices`. Por ejemplo, para encontrar solo la primera letra de vuestro nombre podéis escribir:

````
first_letter = name[0]
print(first_letter)
````

## Diccionarios

Nuestra pequeña lista de buenas lecturas ya parece otra cosa y podemos manupilarla de diferentes formas. Pero imaginemos que empieza a crecer y se vuelve algo dificil de gestionar por ejemplo a la hora de buscar un libro en concreto. Para resolver ese problema Python proporciona una estructura mucho más apropiada que se llama **`diccionario`**. Un diccionario es muy parecido a los diccionarios que tenemos en casa: tienen entradas, o claves (`keys`) y cada entrada tiene un valor (`value`). Por ejemplo:

````
my_dict = {"libro": "conjunto de muchas hojas de papel u otro material semejante que, encuadernadas, forman un volumen",
           "espada": "arma blanca, larga, recta, aguda y cortante, con guarnición y empuñadura",
           "tarta": "pastel grande, de forma generalmente redonda, relleno de frutas, crema, etc., o bien de bizcocho, pasta de almendra y otras clases de masa homogénea"}
````

Fijémonos bien en esta nueva sintaxis. Fijaos que se usan llaves **{** y  **}**. Las entradas están en la parte izquierda y los valores en la derecha. Para encontrar el valor de una entrada lo hacemos utilizando el *'índice'* del diccionario usando el nombre de dicha entrada:

````
description = my_dict["espada"]
print(description)
````

Decimos '`índice`' para usar el mismo término que hemos utilizado anteriormente cuando trabajábamos con listas o cadenas. La única diferencia es que no usamos un número como índice sino la entrada de nuesto diccionario.

Al igual que las listas los diccionarios son **mutables**, lo que significa que podemos añadir y quitar elementos a nuestro antojo.

Deifinamos un diccionario vacío y añadamos algunos libros a él. Los títulos seran las entradas y sus puntuaciones los valores. Veamos como se añade una entrada nueva:

````
good_reads = {}
good_reads["Orgullo y prejuicio"] = 8
good_reads["La naranja mecánica"] = 9
````

Hemos alterado el diccionario de una manera muy parecida a la que hemos visto anteriormente con nuestra lista de libros.
En la lista podíamos acceder a los índices mediante números y en el diccionario podemos acceder directamente con el título del libro lo que hace que sea mucho más cómodo trabajar con ellos cuando nos encontramos con una gran cantidad de entrada.

## Condiciones

### Condiciones simples

En programacion se utilizan mucho las condiciones para que partes del código solo funcionen si se cumplen ciertos requisitos. A continuación podemos ver unos ejemplos:

````
print("2 < 5 =", 2 < 5)
print("3 > 7 =", 3 >= 7)
print("3 == 4 =", 3 == 4)
print("escuela == tarea =", "escuela" == "tarea")
print("Python != perl =", "Python" != "perl")
````

¿Entendéis lo que está pasando? En la segunda parte estamos definiendo nuestras condiciones con **operadores**:
- **>**   Mayor que
- **<**   Menos que
- **==** Igual a

### if, else y elif ("si...", "si no...", "si no, si..."")

Los diccionarios son una estructura de datos mucho mejores para nuesta colección de libros. Sin embargo puede darse el caso de que olvidemos qué libros hemos añadido a nuestra colección. ¿Qué pasa si intentamos obtener la puntuación de un libro que no hemos añadido? (¡Y que esperemos no añadir nunca!)

````
good_reads["Crepúsculo: Luna nueva"]
````

Nos devuelve un **`KeyError`** , o error de clave, que básicamente significa que no tenemos esa entrada en nuestro diccionario. De momento podemos evitar que nuestro programa devuelva el error escribiendo código que muestre "X está en la colección" o, si no se encuentra, "X no está en la colección".

````
book = "La naranja mecánica"
if book in good_reads:
    print(book + " está en la colección")
else:
    print(book + " NO está en la colección")
````

Hay muchas cosas nuevas aquí así que vamos poco a poco.

En primer lugar hemos preguntado si el valor asignado a `book` está en nuestra colección. El `if` evalúa si la condición que hemos escrito es verdadera (`True`) o falsa (`False`). Si escribimos:

````
book in good_reads
````

Como nuestro libro no está en la colección Python devuelve `False`. Hagamos lo mismo con un libro que sabemos que sí está:

````
"La naranja mecánica" in good_reads
````

Lo cual es correcto.

Volvamos al `if` de la expresión. Si la condición que ponemos devuelve `True` entonces Python pasará a ejecutar la siguiente línea y por tanto ejecutará la línea de **`book + " está en la colección"`**. Vamos a probarlo:

````
if "La naranja mecánica" in good_reads:
    print("Encontrado!")
````

````
if book in good_reads:
    print("Encontrado!")
````

Como vemos, el segundo bloque de código no ha hecho nada. Esto se debe a que el valor que hemos asignado a `book` no está en nuestra colección y por tanto la condición no se cumple.

En el primer ejemplo que hemos hecho hay además otra línea con **`else`**. Es el equivalente en español a "si no". Para que quede más claro el funcionamiento podemos traducirlo al español. Este código no funcionará y se muestra solamente como ejemplo.

```
si book in good_reads:
    print(book + " está en la colección")
si no:
    print(book + " NO está en la colección")
```
Es decir: Si el libro está en la colección, muestra que está y si no, muestra que no.

#### ¡Sangría!

Antes de continuar debemos explicar que la disposición de nuestro código no es opcional. En Python el único delimitador de línea que hay son los dos puntos **`:`** y la sangría o indentación del texto. Debemos hacer uso de la sangría siempre en todo nuestro código. Lo convencional es utilizar cuatro espacios.

¿Qué significa esto? Que siempre que usemos los dos puntos (como en nuestro **if** anterior) la siguiente línea debe tener una sangría de 4 espacios mayor que la línea con dos puntos.

A veces querremos comprobar si se cumlen varias condiciones que deberían evaluarse para que hagan cosas diferentes. Para ello en Python usamos **`elif`**. Se usa ded manera parecida al **`if`** y al **`else`** pero solamente podemos usarlo despues de un **`if`**.

En el ejemplo anterior consultábamos si un libro estaba o no en nuestra colección. Podemos hacer lo mismo para partes de una cadena o para elementos de una lista. Por ejemplo podemos probar si la letra *a* está en la palabra *banana*:

````
"a" in "banana"
````

Del mismo modo que la siguiente expresión se evalúa como `False`:

````
"z" in "banana"
````

Probemos una combinación`if-elif-else`:

````
word = "muchas bananas"
if "a" in word:       # si tiene la letra a...
    print(word + " contiene la letra a")
elif "s" in word:     # si no tiene la a, entonces si contiene la s...
    print(word + " contiene la letra s")
else:
    print("¡Vaya palabra!")
````

### and, or, not (y, o, no)

Hasta ahora nuestras condiciones solo contemplaban una expresión. Sin embargo, bastante a menudo querremos probar que varias condiciones se cumplan a la vez. Para esto Python proporciona varias maneras de hacerlo. La primera es con **`and`**.

`and` nos permite evaluar si dos condiciones se dan a la vez. Ambas condiciones deben ser `True`para que la expresión se evalúe como `True`. Veámos cómo funciona:

````
word = "banana"
if "a" in word and "b" in word:
    print("Tanto la a como la b están en " + word)
````

Si una de las condiciones se evalúa como `False`, no se mostrará nada:

````
if "a" in word and "z" in word:
    print("Tanto la a como la z están en " + word)
````

## Bucles

La programación es más útil si podemos realizar una determinada acción sobre una serie de elementos diferentes. Por ejemplo, dada una lista de palabras, nos gustaría saber la longitud de todas las palabras, no sólo de una. Ahora podemos hacerlo recorriendo todos los índices de una lista de palabras e imprimiendo la longitud de las palabras de una en una, ocupando tantas líneas de código como índices tengamos. Huelga decir que esto es bastante engorroso.

Python proporcionala expresión`for` que nos permiten iterar a través de cualquier objeto iterable y realizar acciones sobre sus elementos. El formato básico de un `for` es: 

    for X in iterable:

Podemos mostrar todas las letras de la palabra `banana` de la siguiente manera:

````
for letter in "banana":
    print(letter)
````

El código del bucle se ejecuta tantas veces como letras haya en la cadena. En cada iteración la variable `letter` contiene un valor distinto.

Del mismo modo podemos mostrar todos los elementos de una lista:

````
colors = ["amarillo", "rojo", "verde", "azul", "morado"]
for cualquiercosa in colors:
    print("Este es el color " + cualquiercosa)
````

Y ya que los diccionarios también son objetos iterables, también podemos recorrer nuesta lista de lectura. De este modo iteramos por todas las entradas de nuestro diccionario:

````
for libro in good_reads:
    print(libro)
````

También podemos recorrer los valores además de las entradas de nuestro diccionario:

````
good_reads.items()
````

````
for libro, puntuacion in good_reads.items():
    print(libro + " tiene una puntuación de " + str(puntuacion))
````

`items()` devolverá, con cada iteración (ya que se trata de una lista), un par con la entrada y su valor.

En el ejemplo de arriba, la variable `libro` recorrerá las entradas de nuestro diccionario y la variable `puntuación` recorrerá sus respectivos valores.

Esta es la manera más elegante de recorrer un diccionario, pero podríamos hacerlo de la siguiente manera también:

````
for book in good_reads:
    print(book, "has score", good_reads[book])
````

Ahora queremos eliminar todos los signos de puntuación de un texto, no sólo los puntos y las comas. Escribiremos una función llamada `eliminar_puntuacion` que elimina todos los signos de puntuación de un texto. Una vez más, hay muchas maneras en las que podemos escribir esta función. Te mostraremos dos de ellas. La primera estrategia es llamar repetidamente a `reemplazar` en la misma cadena cada vez, reemplazando un carácter de puntuación diferente por una cadena vacía. 

````
def eliminar_puntuacion(texto):
    signos_puntuacion = '!¡@#$%^&*()_-+={}[]:;"\'|<>,.¿?/~`'
    for signo in signos_puntuacion:
        texto = texto.replace(signo, "")
    return texto

texto_corto = "Las comas, parece ser, están sobreestimadas. Y, los puntos, ¡más todavía!."
print(eliminar_puntuacion(texto_corto))
````

La segunda estrategia que seguiremos es mostraos que podemos lograr el mismo resultado sin usar la función `replace` que viene incorporada en Python. Recordemos que una cadena está formada por caracteres. Podemos hacer un bucle sobre una cadena accediendo a cada carácter por separado. Cada vez que encontramos un signo de puntuación saltamos al siguiente carácter.

````
def eliminar_puntuacion_2(texto):
    signos_puntuacion = '¡!@#$%^&*()_-+={}[]:;"\'|<>,.¿?/~`'
    texto_limpio = ""
    for caracter in texto:
        if caracter not in signos_puntuacion:
            texto_limpio += caracter
    return texto_limpio

texto_corto = "Las comas, parece ser, están sobreestimadas. Y, los puntos, ¡más todavía!."
print(eliminar_puntuacion_2(texto_corto))
````

Para escribir algo en un archivo abrimos, al igual que hicimos al leerlo, un "TextIOWrapper" que puede ser visto como una conexión con el archivo `primera-salida.txt`. La diferencia con abrir un archivo para leer es el **modo** con el que se abre la conexión. Aquí el modo dice `w`, que significa **"abrir el archivo para escribir"**. Para abrir un archivo de lectura, establecemos el modo en `r`. Sin embargo, como esta es la configuración por defecto de Python, podemos omitirla.

````
from IPython.display import HTML
import base64

def create_download_link( file, filename , title = "Descargar archivo"):  
    f = open(filename)
    file = f.read()
    f.close()
    b64 = base64.b64encode(file.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

create_download_link(archivo_salida, "primera-salida.txt")
````
